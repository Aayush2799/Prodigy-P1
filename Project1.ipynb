Importing Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Loading the data

```python
df = pd.read_csv('/content/drive/MyDrive/Prodigy/P1/API_SP.POP.TOTL_DS2_en_csv_v2_2001050.csv',skiprows=4)
```

```python
df.head()
```

Selecting relevant columns and renaming them for better convenience

```python
df_2011 = df[['Country Name', '2011']].dropna()
```

```python
df_2011.columns = ['Country', 'Population']
```

```python
df_2011 = df_2011.sort_values(by='Population', ascending=False)
```

### Plotting Bar Graph

```python
plt.figure(figsize=(18, 9))
plt.bar(df_2011['Country'][:27], df_2011['Population'][:27])
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 20 Most Populated Countries in 2022')
plt.xticks(rotation=90)
plt.show()
```
